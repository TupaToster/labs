```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constants
R_0 = 0.3 #ohm
R_u = 20 * 10**3 #ohm
C_u = 20 * 10**-6 #farad
```

```python
#Sample one
#Пермаллой Fe-Ni НП50
N_0 = 40 #витков
N_u = 200 #витков
S = 3.8 #cm^2
round_len = 24 #cm, 2 * pi * R

K_x = 0.02 / 5
K_y = 0.1 / 5

X_s = [14,13,12,11,10,9,8,7 ]
Y_s = [
8,7,7,7,6,4,1,0
]

plt.figure (figsize = (16, 9), dpi = 200)

plt.grid ()

plt.xlabel (r'$X_s$, Дел')
plt.ylabel (r'$Y_s$, Дел')

plt.plot (X_s, Y_s, color = 'black', linestyle = '-', marker = '+', linewidth = 0.5, markersize = 6)

plt.legend ()

plt.savefig ('graph1.png')
plt.show ()

H = (K_x * N_0) / (round_len * R_0)
B = (K_y * R_u * C_u) / (S * N_u)

H, B
#data in table sheet1
```
